## Problem 1. 
Caesar Cipher - given a string and an integer $n$, return the string with each letter shifted by n positions in the alphabet. Non-letter characters should remain unchanged. The shift should wrap around the alphabet (e.g., shifting 'z' by 1 should result in 'a').

```python
def caesar(text: str, shift: int) -> str:
    result = ""

    for char in text:
        if char.isalpha():
            base = ord("A") if char.isupper() else ord("a")
            shifted = (ord(char) - base + shift) % 26 + base
            result += chr(shifted)
        else:
            result += char

    return result
```

```python
text1 = "Hello, there"
text1_encrypted = caesar(text1, 2)
print(f"Original: {text1} -> Encrypted: {text1_encrypted}")
```

### Problem 1.2
Given a encrypted string using Caesar Cipher return all possible decryptions by trying all shifts from 1 to 26.

```python
def decrypt(text: str) -> list[str]:
    return [caesar(text, i) for i in range(1, 26 + 1)]
```

### Problem 2.1
Reading from a file: Write a function that takes a filename as input and returns the lines of the file as a list of strings without newline characters.

```python
def read_file(filename: str) -> list[str]:
    with open(filename, "r") as file:
        return [line.strip() for line in file.readlines()]
```

```python
read_file("08_files/sample.txt")
```

### Problem 2.2
Writing to a file: Write a function that takes a filename and a list of strings as input, and writes each string to the file on a new line.


```python
def write_file(filename: str, lines: list[str]) -> None:
    with open(filename, "w") as file:
        for line in lines:
            file.write(line + "\n")
```

```python
lines = ["Line one", "Line two", "Line three"]
write_file("08_files/output.txt", lines)
```
