## Problem 1:
Given two texts $A$ and $B$, write a function to determine whether $A = B$.

```python
def are_texts_equal(n, m):
    if len(n) != len(m):
        return False

    for i in range(len(n) - 1):
        if n[i] != m[i]:
            return False

    return True
```

```python
print(are_texts_equal("ala", "ala"))  # True
print(are_texts_equal("ala", "ola"))  # False
print(are_texts_equal("ala", "alaa"))  # False
```

## Problem 2:
Given a text $t$, determine whether it is a palindrome (i.e., it reads the same forwards and backwards).

```python
def is_palidrome(t):
    return t == t[::-1]
```

```python
print(is_palidrome("ala"))  # True
print(is_palidrome("ola"))  # False
print(is_palidrome("abba"))  # True
print(is_palidrome("abcd"))  # False
print(is_palidrome("a"))  # True
print(is_palidrome(""))  # True
```

## Problem 3:
Given a text $t$, print all palindromic substrings of $t$ which are at least two characters long.

```python
def print_all_palidromes(t):
    n = len(t)
    for i in range(n):
        for j in range(i + 2, n + 1):
            substring = t[i:j]
            if is_palidrome(substring):
                print(substring, end=", ")
    print()
```

```python
print_all_palidromes("ababa")  # ababa, aba, bab, aba
print_all_palidromes("abcd")  # (no output)
print_all_palidromes("aabbbaa")  # aabbbaa, abbba, bb, bbb, aa
```
